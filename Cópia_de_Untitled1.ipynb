```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# quantas linhas e colunas?
print(df.shape)

# primeiras linhas
print(df.head())

# estatísticas básicas
print(df.describe())

# valores nulos por coluna
print(df.isnull().sum())
```

```python
# quantos sobreviveram?
print(df['Survived'].value_counts())

# taxa de sobrevivência por classe
print(df.groupby('Pclass')['Survived'].mean().round(2))

# média de idade dos sobreviventes vs não sobreviventes
print(df.groupby('Survived')['Age'].mean().round(1))

# filtrar só mulheres que sobreviveram
sobreviventes_f = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print(len(sobreviventes_f))
```

```python
import matplotlib.pyplot as plt

# sobrevivência por classe — gráfico de barras
df.groupby('Pclass')['Survived'].mean().plot(kind='bar', color='steelblue')
plt.title('Taxa de Sobrevivência por Classe')
plt.xlabel('Classe')
plt.ylabel('Taxa')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# distribuição de idades
df['Age'].dropna().hist(bins=20, color='steelblue', edgecolor='white')
plt.title('Distribuição de Idades')
plt.show()
```
