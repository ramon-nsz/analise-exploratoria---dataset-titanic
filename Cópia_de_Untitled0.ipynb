```python
# carregar e explorar dados
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print(df.shape)           # quantas linhas e colunas
print(df.head())          # primeiras linhas
print(df.describe())      # estatísticas básicas
print(df.isnull().sum())  # valores nulos por coluna
```

```python
# filtrar e agrupar
print(df['Survived'].value_counts())
print(df.groupby('Pclass')['Survived'].mean())
print(df['Age'].mean())
```

```python
# Bloco 3 — gráfico simples
import matplotlib.pyplot as plt

df['Age'].dropna().hist(bins=20)
plt.title('Distribuição de Idade')
plt.xlabel('Idade')
plt.show()
```
